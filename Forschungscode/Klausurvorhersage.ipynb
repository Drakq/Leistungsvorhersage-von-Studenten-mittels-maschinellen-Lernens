```python
#Alte Notenumformung
def toNote(punkte):    
    schritt = (np.array(punkte)*(20/100)).astype(int)
    noten = (22-schritt)*(1/3)
    noten[noten<1.0] = 1.0
    noten[noten>4.0] = 5.0
    noten = np.round(noten,1)
    return noten

#Neue Notenumformung nach Jens (viel langsamer, aber ey...)
def toNote2(erreichtePunkte):
    punkte = np.floor(erreichtePunkte*(20/100))
    if punkte < 10: return 5.0
    elif punkte >= 19: return 1.0
    else: return {10: 4.0,
                  11: 3.7,
                  12: 3.3,
                  13: 3.0,
                  14: 2.7,
                  15: 2.3,
                  16: 2.0,
                  17: 1.7,
                  18: 1.3}[punkte]
toNote2 = np.vectorize(toNote2)

df_klausur["note-korrigiert"] = toNote(df_klausur["klausur-punkte"])

print("Notenschnitt:", round(df_klausur["note"].sum()/len(df_klausur["note"]),2))
print("Notenschnitt(korrigiert):", round(df_klausur["note-korrigiert"].sum()/len(df_klausur["note-korrigiert"]),2))

plt.rcParams['font.size'] = 16
plt.figure(figsize=(15, 10))
plt.grid(color="black", linestyle="--", linewidth=1, alpha=0.15)
plt.scatter(df_klausur["klausur-punkte"][df_klausur["fach"]==0], df_klausur["note-korrigiert"][df_klausur["fach"]==0], s=100, c="b", label="Informatik", alpha=0.5)
plt.scatter(df_klausur["klausur-punkte"][df_klausur["fach"]==1], df_klausur["note-korrigiert"][df_klausur["fach"]==1], s=100, c="cyan", label="Mathematik", alpha=0.5)
plt.scatter(df_klausur["klausur-punkte"][df_klausur["fach"]==2], df_klausur["note-korrigiert"][df_klausur["fach"]==2], s=100, c="lime", label="Physik", alpha=0.5)
plt.scatter(df_klausur["klausur-punkte"][df_klausur["fach"]==3], df_klausur["note-korrigiert"][df_klausur["fach"]==3], s=100, c="r", label="Informationswissenschaften", alpha=0.5)
plt.scatter(df_klausur["klausur-punkte"][df_klausur["fach"]==4], df_klausur["note-korrigiert"][df_klausur["fach"]==4], s=100, c="m", label="Andere", alpha=0.5)
plt.xlabel("Klausurpunkte", fontsize=16)
plt.ylabel("Note", fontsize=16)
plt.yticks((1.0, 1.3, 1.7, 2.0, 2.3, 2.7, 3.0, 3.3, 3.7, 4.0, 5.0))
plt.legend(loc="lower left", fontsize=16)

plt.tight_layout(pad=0.1)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import r2_score

def regression(einflussdaten, zieldaten, maxGrad=10):
    gridSearch = pd.DataFrame(columns=["Grad","MAE", "MSE", "Adj. R2", "Schätzer"])
    for grad in range(1, maxGrad+1):
        vorhersage = np.array([])
        real = np.array([])
        poly = PolynomialFeatures(grad)
        for train, test in LeaveOneOut().split(einflussdaten):
            X_train, X_test, y_train, y_test = einflussdaten.iloc[train], einflussdaten.iloc[test], zieldaten.iloc[train], float(zieldaten.iloc[test])
            transform_X_train = poly.fit_transform(X_train)
            linReg = LinearRegression()
            linReg.fit(transform_X_train, y_train)
            
            transform_X_test = poly.fit_transform(X_test)
            vorhersage = np.append(vorhersage, [linReg.predict(transform_X_test)[0]])
            real = np.append(real, [y_test])
        
        mae = round(np.mean(abs(vorhersage-real)),3)
        mse = round(np.mean((vorhersage-real)**2),3)
        r2 = r2_score(real, vorhersage)
        n, p = einflussdaten.shape
        r2_adj = round(1-(1-r2)*(n-1)/(n-p-1),4)
        
        transform_X_train = poly.fit_transform(einflussdaten)
        linReg = LinearRegression()
        linReg.fit(transform_X_train, zieldaten)
            
        gridSearch.loc[grad] = [grad, mae, mse, r2_adj, linReg]
    return gridSearch

regression_einfach = regression(df_klausur[["summe-uebungs-punkte"]], df_klausur["klausur-punkte"], maxGrad=10)
display(regression_einfach)

#for i in range(5,6):
#    plt.rcParams['font.size'] = 16
#    plt.figure(figsize=(15, 10))
#    plt.scatter(df_klausur[["summe-uebungs-punkte"]], df_klausur["klausur-punkte"], s=500, c="b", alpha=2/3)
#    x = np.arange(min(df_klausur["summe-uebungs-punkte"])-10, max(df_klausur["summe-uebungs-punkte"])+10, 1).reshape(-1,1)
#    plt.plot(x, regression_einfach["Schätzer"][i].predict(PolynomialFeatures(i).fit_transform(x)), "r-", linewidth=3, label="Schätzerfunktion")
#    plt.title("Grad "+str(i))
#    plt.xlabel("Übungspunkte", fontsize=24)
#    plt.ylabel("Klausurpunkte", fontsize=24)
#    plt.legend(loc="upper left", fontsize=23)
#    plt.tight_layout(pad=0.1)
#    #plt.savefig("LaTeX/fig/3.1_overfitting.png", transparent=True)
#    plt.show()
```

## kNearestNeighbor

```python
from sklearn.neighbors import KNeighborsRegressor

def knn(einflussdaten, zieldaten, k=1):
    vorhersage = np.array([])
    real = np.array([])
    for train, test in LeaveOneOut().split(einflussdaten):
        X_train, X_test, y_train, y_test = einflussdaten.iloc[train], einflussdaten.iloc[test], zieldaten.iloc[train], float(zieldaten.iloc[test])
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(X_train, y_train)
        
        vorhersage = np.append(vorhersage, knn.predict(X_test))
        real = np.append(real, y_test)
    mae = round(np.mean(abs(vorhersage-real)),3)
    mse = round(np.mean((vorhersage-real)**2),3)
    r2 = r2_score(real, vorhersage)
    n, p = einflussdaten.shape
    r2_adj = round(1-(1-r2)*(n-1)/(n-p-1),4)
    print("Bestes k:", k, "MAE:", mae, "MSE:", mse, "R2:", r2_adj)

knn(df_klausur_reg[["summe-uebungs-punkte","uebungspunkte-pro-blatt"]], df_klausur_reg["klausur-punkte"], k=11)
```
